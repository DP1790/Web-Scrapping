```python
url = 'https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos'
driver.get(url)
```

```python
# Lists
Rank = [] 
Name = [] 
Artist = [] 
Upload_date = [] 
Views = []


try: # Scrapping Rank
    for i in driver.find_elements_by_xpath("//div/table[3]/tbody/tr/td[1]"):
        Rank.append(i.text)
except NoSuchElementException as e:
    print("No such element present")
    
try: # Scrapping Name
    for i in driver.find_elements_by_xpath("//div/table[3]/tbody/tr/td[2]"):
        Name.append(i.text)
except NoSuchElementException as e:
    print("No such element present")
    
try: # Scrapping Artist
    for i in driver.find_elements_by_xpath("//div/table[3]/tbody/tr/td[3]"):
        Artist.append(i.text)
except NoSuchElementException as e:
    print("No such element present")
    
try: # Scrapping Upload date
    for i in driver.find_elements_by_xpath("//div/table[3]/tbody/tr/td[4]"):
        Upload_date.append(i.text)
except NoSuchElementException as e:
    print("No such element present")
    
try: # Scrapping Views
    for i in driver.find_elements_by_xpath("//div/table[3]/tbody/tr/td[5]"):
        Views.append(i.text)
except NoSuchElementException as e:
    print("No such element present")

```

```python
#creating data frame 
youtube_videos=pd.DataFrame({"Rank":Rank,"Name":Name,"Artist":Artist,"Uplode_Date":Upload_date,"Views":Views})
```

```python
# Lists
name = []
year = []
genre = []
time = []
rat = []
vot = []

try: # Scraping Name
    for i in driver.find_elements_by_xpath("//h3[@class='lister-item-header']/a"):
        name.append(i.text)
except NoSuchElementException as e:
    print("-")

    
try: # Scraping year span
    for i in driver.find_elements_by_xpath("//span[@class='lister-item-year text-muted unbold']"):
        year.append(i.text)
except NoSuchElementException as e:
    print("-")
        
try: # Scraping genre
    for i in driver.find_elements_by_xpath("//div[@class='lister-item-content']/p/span[5]"):
        genre.append(i.text)
except NoSuchElementException as e:
    print("-")
    
try: # Scraping Run time
    for i in driver.find_elements_by_xpath("//div[@class='lister-item-content']/p/span[3]"):
        time.append(i.text)
except NoSuchElementException as e:
    print("-") 
    
try: # Scraping Ratings
    for i in driver.find_elements_by_xpath("//div[@class='ipl-rating-star small']/span[2]"):
        rat.append(i.text)
except NoSuchElementException as e:
    print("-")
    
try: # Scraping votes
    for i in driver.find_elements_by_xpath("//div[@class='lister-item-content']/p[4]/span[2]"):
        vot.append(i.text)
except NoSuchElementException as e:
    print("-")
```

```python
#creating data frame 
imdb = pd.DataFrame({'Name' : name,
    'Year Span' : year,
    'Genre' : genre,
    'Run Time': time,
    'Ratings' : rat,
    'Votes': vot})
imdb.head()
```

```python
alldata = driver.find_element_by_xpath('/html/body/table[1]/tbody/tr/td[2]/span[2]/a/font/b').click()
```

```python
# lists
name = []
data = []
task = []
attr = []
ins = []
no_attr = []
year = []
   
try: # Scraping Name
    for i in driver.find_elements_by_xpath("//table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[1]"):
        name.append(i.text)
except NoSuchElementException as e:
    print("-")

    
try: # Scraping data type
    for i in driver.find_elements_by_xpath("//table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[2]"):
        data.append(i.text)
except NoSuchElementException as e:
    print("-")
        
try: # Scraping task
    for i in driver.find_elements_by_xpath("//table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[3]"):
        task.append(i.text)
except NoSuchElementException as e:
    print("-")
    
try: # Scraping attribute type
    for i in driver.find_elements_by_xpath("//table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[4]"):
        attr.append(i.text)
except NoSuchElementException as e:
    print("-") 
    
try: # Scraping no of instance
    for i in driver.find_elements_by_xpath("//table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[5]"):
        ins.append(i.text)
except NoSuchElementException as e:
    print("-")
    
try: # Scraping no of attribute
    for i in driver.find_elements_by_xpath("//table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[6]"):
        no_attr.append(i.text)
except NoSuchElementException as e:
    print("-")
    
try: # Scraping no of attribute
    for i in driver.find_elements_by_xpath("//table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[7]"):
        year.append(i.text)
except NoSuchElementException as e:
    print("-")    
```

```python
#creating data frame 
uci = pd.DataFrame({'Dataset Name': name[1:],
    'Data Type': data[1:],
    'Task' : task[1:],
    'Attribute Type': attr[1:],
    'No of Instances' :ins[1:],
    'No of Attribute': no_attr[1:],
    'Year': year[1:]
    })
uci.head()
```
