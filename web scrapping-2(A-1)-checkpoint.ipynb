```python
from bs4 import BeautifulSoup
import requests
```

```python
page = requests.get("https://bookpage.com/reviews?page=56")
page
```

```python
soup = BeautifulSoup(page.content)
soup
```

```python
books_data = soup.find_all('h4', class_= 'italic')
books_data
```

```python
books_data[0].text.replace('\n','')
```

```python
book_name = []
for i in books_data:
    book_name.append(i.text.replace('\n',''))
book_name[:5]
```

```python
books_data1 = soup.find_all('p', class_= 'sans bold')
books_data1
```

```python
books_data1[0].text.replace('\n','')
```

```python
author_name = []
for a in books_data1:
    author_name.append(a.text.replace('\n',''))
author_name[:5]
```

```python
books_data2 = soup.find_all('p', class_= 'genre-links hidden-phone')
books_data2
```

```python
books_data2[0].text.replace('\n','')
```

```python
genre = []
for g in books_data2:
    genre.append(g.text.replace('\n',''))
genre[:5]
```

```python
books_data3 = soup.find_all('p', class_= 'excerpt')
books_data3
```

```python
book_review = []
for r in books_data3:
    book_review.append(r.text.replace('\n',''))
book_review[:5]
```

```python
page1 = requests.get("https://www.icc-cricket.com/rankings/mens/team-rankings/odi")
page1
```

```python
soup1 = BeautifulSoup(page1.content)
soup1
```

```python
cricket_data = soup1.find_all('span', class_= "u-hide-phablet")
cricket_data[0].text.replace('\n','')
```

```python
teams_name = []
for c in cricket_data:
    teams_name.append(c.text.replace('\n',''))
teams_name[:10]
```

```python
Matches = [soup1.find("td", class_ = "rankings-block__banner--matches").text]
Points = [soup1.find("td", class_ = 'rankings-block__banner--points').text]


for m,j in enumerate(soup1.find_all("td", class_= "table-body__cell u-center-text")[:18]):
    if m%2 == 0:
        Matches.append(j.text)
    else:
        Points.append(j.text)
```

```python
print("Matches: ",Matches)
print("Points :",Points)
```

```python
Ratings = [soup1.find("td", class_ = "rankings-block__banner--rating u-text-right").text.replace('\n','').strip()]
for r in soup1.find_all("td", class_= "table-body__cell u-text-right rating")[:9]:
    Ratings.append(r.text)
```

```python
print("Ratings :",Ratings)
```

```python
import pandas as pd
top10_cricket = pd.DataFrame()
top10_cricket['Teams'] = teams_name[:10]
top10_cricket['Matches'] = Matches[:10]
top10_cricket['Points'] = Points[:10]
top10_cricket['Ratings'] = Ratings[:10]

top10_cricket
```
