```python
# Fetching urls to open each page.
urls =[]
for i in range(0,3): #upto 3 pages
    for x in driver.find_elements_by_xpath("//a[@class='a-link-normal a-text-normal']"):
        urls.append(x.get_attribute('href'))
    driver.find_element_by_xpath("//li[@class='a-last']/a").click()
    time.sleep(3)
```

```python
#Lists
brand=[]
prod_name=[]
rating=[]
rating_num=[]
price= []
ret_ex=[]
exp_del = []
availability=[]
prod_url=[]
ot_det = []
```

```python
# Scrapping the all details

for i in urls[:150]:
    driver.get(i)
    time.sleep(3)
    
    #fetching brand name
    try:
        brand_name=driver.find_element_by_xpath("//td[@class='a-span9']/span")
        brand.append(brand_name.text)
    except NoSuchElementException:
        brand.append('-')    
        
        
      #fetching product name
    try:
        product=driver.find_element_by_id("productTitle")
        prod_name.append(product.text)
    except NoSuchElementException:
        prod_name.append('-') 
        
        
      #fetching rating
    try:
        ratestar=driver.find_element_by_xpath("//span[@class='a-size-base a-nowrap']//span")
        rating.append(ratestar.text)
    except NoSuchElementException:
        rating.append("No Rating")
        
        
        
      #fetching rating number
    try:
        ratenum=driver.find_element_by_id("acrCustomerReviewText")
        rating_num.append(ratenum.text)
    except NoSuchElementException:
        rating_num.append('-') 
        
        
        
      #fetching Price
    try:
        prod_price=driver.find_element_by_id("priceblock_ourprice")
        price.append(prod_price.text)
    except NoSuchElementException:
        price.append('-') 
        
        
        
     #fetching Return/Exchange
    try:
        ret_exch=driver.find_element_by_xpath("//a[@class='a-size-small a-link-normal a-text-normal']")
        ret_ex.append(ret_exch.get_attribute("href"))
    except NoSuchElementException:
        ret_ex.append('-') 
        
     #fetching expected delivery
    try:
        ex_de=driver.find_element_by_xpath("//div[@class='a-section a-spacing-mini']/b")
        exp_del.append(ex_de.text)
    except NoSuchElementException:
        exp_del.append('-')
           
        
       
     #fetching Availability
    try:
        availa=driver.find_element_by_id("availability")
        availability.append(availa.text)
    except NoSuchElementException:
        availability.append('-') 
        
        
        
           
     #fetching Product URL
    try:
        pro_url=driver.find_element_by_xpath("//div[@class='a-section a-spacing-none icon-content']/a")
        prod_url.append(pro_url.get_attribute("href"))
    except NoSuchElementException:
        prod_url.append('-')
        
     #fetching Other details
    try:
        othr_det=driver.find_element_by_xpath("//ul[@class='a-unordered-list a-vertical a-spacing-mini']/li/span")
        ot_det.append(othr_det.text)
    except NoSuchElementException:
        ot_det.append('-') 
```

```python
# creating dataframe.
guitar_details =pd.DataFrame({})
guitar_details['Brand Name'] = brand
guitar_details['Product Name'] = prod_name
guitar_details['Rating'] = rating
guitar_details['Rating Number'] = rating_num
guitar_details['Price'] = price
guitar_details['Return/Exchange'] = ret_ex
guitar_details['Expected Delivery'] = exp_del
guitar_details['Availability'] = availability
guitar_details['Product URL'] = prod_url
guitar_details['Other Details'] = ot_det
```

```python
# Make empty lists
urls = []
name =[]
distance = []
ratings = []
total_reviews = []
price = []
private = []
dorms = []
facilities = []
property_description = []


for i in driver.find_elements_by_xpath("//h2[@class='title title-6']/a"):
    urls.append(i.get_attribute('href'))

for i in driver.find_elements_by_xpath("//span[@class='description']"):
    distance.append(i.text)
try:    
    for i in driver.find_elements_by_xpath("//div[@class='score orange big']"):
        ratings.append(i.text)
except NoSuchElementException:
    ratings.append('--')
    
for i in driver.find_elements_by_xpath("//div[@class='reviews']"):
    total_reviews.append(i.text)

for i in driver.find_elements_by_xpath('//div[@class="price-col"]'):
    price.append(i.text)

for i in price:
    if 'Privates' in i:
        private.append(i.replace('/n',''))
    else:
        dorms.append(i.replace('/n',''))

for url in urls:
    driver.get(url)
    time.sleep(4)
    for i in driver.find_elements_by_xpath("//div[@class='title-2']"):
        name.append(i.text)
        
    for i in driver.find_elements_by_xpath('//ul[@class="facilities"]'):
        facilities.append(i.text)
    
    for i in driver.find_elements_by_xpath("//div[@class='content collapse-content']"):
        property_description.append(i.text)        
```
